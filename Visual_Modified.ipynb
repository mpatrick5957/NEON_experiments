```python
#Import Libraries
%matplotlib inline

import os
import sys
import time
import datetime

import numpy as np
import pandas as pd
import xarray as xr

from glob import glob
from os.path import join, expanduser

import matplotlib
import matplotlib.pyplot as plt

from scipy import stats

from neon_utils import download_eval_files
```

```python
#Specify the year below
year = "2018"
neon_site = "BART"
```

```python
tree_path = "/home/user/archive/BART_Tree.transient/lnd/hist/"
tree_files = sorted(glob(join(tree_path,neon_site+".transient.clm2.h1."+year+"*.nc")))

print("All Simulation files: [", len(tree_files), "files]")
```

```python
grass_path = "/home/user/archive/BART_grass.transient/lnd/hist/"
grass_files = sorted(glob(join(grass_path,neon_site+".transient.clm2.h1."+year+"*.nc")))

print("All Simulation files: [", len(grass_files), "files]")
```

```python
start = time.time()

ds_tree = xr.open_mfdataset(tree_files, decode_times=True, combine='by_coords')

end = time.time()
print("Reading all simulation files took:", end-start, "seconds.")
```

```python
start = time.time()

ds_grass = xr.open_mfdataset(grass_files, decode_times=True, combine='by_coords')

end = time.time()
print("Reading all simulation files took:", end-start, "seconds.")
```

```python
ds_tree['GPP'].plot(label = "tree")
ds_grass['GPP'].plot(label = "grass")
plt.legend()
```

```python
(ds_tree['GPP']-ds_grass['GPP']).plot(label = "tree-grass")

plt.legend()
```
